# Setup process of DuckDB
Run this notebook to create DuckDB if we need to re-create the database
*If cloned from GitHub, this is already created - however; .venv is not guaranteed to be compatible. 

```python
import os
import duckdb

# Construct the path to the data path
db_path = os.path.join("..", "data", "wikipedia.db") 

# create database wikipedia
conn = duckdb.connect(db_path)

# create schema
conn.execute("CREATE SCHEMA IF NOT EXISTS iv")

create_table_sql = """
        CREATE TABLE IF NOT EXISTS iv.wiki_edits (
            type VARCHAR,
            title VARCHAR,
            user VARCHAR,
            userid BIGINT,
            timestamp TIMESTAMP,
            comment VARCHAR
        )
        """
conn.execute(create_table_sql)

conn.close()
```
